# Regression

Tracking Parkinson's disease symptom progression often uses the unified Parkinson's disease rating scale (UPDRS) that requires the patient's presence in clinic, and time-consuming physical examinations by trained medical staff. Thus, symptom monitoring is costly and logistically inconvenient for patient and clinical staff alike, also hindering recruitment for future large-scale clinical trials. 

In 2010 a data set was published [1] that was composed of a range of biomedical voice measurements from 42 people with early-stage Parkinson's disease recruited to a six-month trial of a telemonitoring device for remote symptom progression monitoring. The recordings were automatically captured in the patient's homes. 


> [1] A. Tsanas, M. A. Little, P. E. McSharry, L. O. Ramig (2010),
*Accurate telemonitoring of Parkinson's disease progression by non-invasive 
speech tests*,
IEEE Transactions on Biomedical Engineering;57(4):884-93.

Let's see if we can create an accurate predictor of UPDRS from the voice measurements. Load the data set, the file is called "parkinsons_updrs.csv".

```python
import pandas as pd

data = pd.read_csv("parkinsons_updrs.csv")
```

What is the shape of the data set?

```python
data.shape
```

The target column for the regression analysis is 'motor_UPDRS'. Plot the distrubtion of this target column using the python `seaborn` package.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,6))
sns.distplot(data['motor_UPDRS'])
plt.show()
```

Remove the "motor_UPDRS" column from the data set and assign it to a variable called `target`.

```python
target = data.pop('motor_UPDRS')
```

Investigate the distribution of the features by plotting them as boxplots. What do you see?

```python
plt.figure(figsize=(12,6))
data.boxplot(vert=False,flierprops=dict(marker='o', markersize=5))
plt.show()
```

Not very much, except that some features have much larger scales than others. To obtain better boxplots we can minmax scale the features. This does not change the distribution of the features.

```python
from sklearn.preprocessing import MinMaxScaler

data_norm = pd.DataFrame(MinMaxScaler().fit_transform(data),columns=data.columns.values)

plt.figure(figsize=(12,6))
data_norm.boxplot(vert=False,flierprops=dict(marker='o', markersize=5))
plt.show()
```

Now we can see the boxplots much better. Plot a histogram of the feature 'Shimmer'.

```python
plt.figure(figsize=(6,6))
sns.distplot(data_norm['Shimmer'])
plt.show()
```

Notice how this distribution has a long tail on the right. It is sometimes useful to transform the feature values to remove this tail. Let's see if this is true for our data set. Train a linear model on this data using the `LinearRegression` module in scikit-learn. What is the $R^2$ performance of the regression model on the data set (so how well does the linear model fit the full data set)?

```python
from sklearn import metrics
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(data_norm,target)

print "R-squared = %f" % metrics.r2_score(target,model.predict(data_norm))
```

Now we will transform some of the features to remove their tails and see if there is a difference. One way to transform a feature with a long tail is to use the $log_2$ function.

```python
import numpy as np

Shimmer_log2 = np.log2(data_norm['Shimmer']+0.01)
```

Notice how we added 0.01 to the column before taking the log to prevent taking the log of zero. Plot a histogram of the variable `Shimmer_log2` using `distplot()`.

```python
plt.figure(figsize=(6,6))
sns.distplot(Shimmer_log2)
plt.show()
```

As you can see most of the tail has mostly disappeared. There are more columns with long tails. Create a python variable `data_norm_log2` that is a copy of `data_norm`. Use a `for` loop to $log_2$ transform the feature values in `data_norm_log2` listed in the python list `to_log_scale` declared below.

```python
to_log_scale= ['Jitter(%)', 'Jitter(Abs)', 'Jitter:RAP', 'Jitter:PPQ5', 'Jitter:DDP', 'Shimmer(dB)',
       'Shimmer:APQ3', 'Shimmer:APQ5', 'Shimmer:APQ11', 'Shimmer:DDA','Shimmer','NHR','PPE']

data_norm_log2 = data_norm.copy()

for c in to_log_scale:
    data_norm_log2[c] = np.log2(data_norm_log2[c]+0.01)
```

Make the boxplots of the features in `data_norm_log2`.

```python
plt.figure(figsize=(12,6))
data_norm_log2.boxplot(vert=False,flierprops=dict(marker='o', markersize=5))
plt.show()
```

Now use the `StandardScaler` module in the sklearn package standardize the features in `data_norm_log2` and plot the boxplots of the features.

```python
from sklearn.preprocessing import StandardScaler

data_norm_log2 = pd.DataFrame(StandardScaler().fit_transform(data_norm_log2),columns=data.columns.values)

plt.figure(figsize=(12,6))
data_norm_log2.boxplot(vert=False,flierprops=dict(marker='o', markersize=5))
plt.show()
```

Fit the model `LinearRegression` model again. How well does it fit this data set now?

```python
model.fit(data_norm_log2,target)

print "R-squared = %f" % metrics.r2_score(target,model.predict(data_norm_log2))
```

Maybe we can improve the fit by assuming a non-linear relationship between the features and the target. To add polynomial features of degree $d$ we can use the `PolynomialFeatures` module in sklearn as follows:

```python
from sklearn.preprocessing import PolynomialFeatures

d=3

polynomial_features = PolynomialFeatures(degree=d)
data_norm_log2_poly=pd.DataFrame(polynomial_features.fit_transform(data_norm_log2))
```

Do the same for the `data_norm` data set. Store the result in variable `data_norm_poly`.

```python
data_norm_poly=pd.DataFrame(polynomial_features.fit_transform(data_norm))
```

How many features are there now?

```python
data_norm_poly.shape[1]
```

Now investigate the first 100 features in `data_norm_log2_poly` using a boxplot. What do you see?

```python
plt.figure(figsize=(12,14))
data_norm_log2_poly.iloc[:,:100].boxplot(vert=False,flierprops=dict(marker='o', markersize=5))
plt.show()
```

Do the same for `data_norm_poly`.

```python
plt.figure(figsize=(12,14))
data_norm_poly.iloc[:,:100].boxplot(vert=False,flierprops=dict(marker='o', markersize=5))
plt.show()
```

For both data sets `data_norm_poly` and `data_norm_log2_poly` train a linear model and compute the $R^2$ performance.

```python
model.fit(data_norm_poly,target)
print "R-squared = %f" % metrics.r2_score(target,model.predict(data_norm_poly))
model.fit(data_norm_log2_poly,target)
print "R-squared = %f" % metrics.r2_score(target,model.predict(data_norm_log2_poly))
```

We can investigate the predictions further by looking at the correlation between the target 'motor_UPDRS' and the predictions using a scatterplot:

```python
import seaborn as sns

model.fit(data_norm_poly,target)

tmp = pd.DataFrame()
tmp['target'] = target
tmp['prediction'] = model.predict(data_norm_poly)

sns.lmplot(x="target", y="prediction", data=tmp, size=7.5, scatter_kws={"s": 10})
plt.show()
```

Make the same plot for the predictions computed for `data_norm_log2_poly'.

```python
model.fit(data_norm_log2_poly,target)

tmp = pd.DataFrame()
tmp['target'] = target
tmp['prediction'] = model.predict(data_norm_log2_poly)

sns.lmplot(x="target", y="prediction", data=tmp, size=7.5, scatter_kws={"s": 10})
plt.show()
```

Now we will split the data sets into a train and a test at random by creating a list "folds" that contains random values zero and one. 

```python
import numpy as np

np.random.seed(1234)

folds = np.random.randint(2,size=len(data_norm_poly))
```

What does the following code do?

```python
d=3

polynomial_features = PolynomialFeatures(degree=d)
data_norm_poly=pd.DataFrame(polynomial_features.fit_transform(data_norm))

model.fit(data_norm_poly[folds==0],target[folds==0])

print "R-squared = %f" % metrics.r2_score(target[folds==0],model.predict(data_norm_poly[folds==0]))
```

what is the $R^2$ performance on the test set?

```python
print "R-squared = %f" % metrics.r2_score(target[folds==1],model.predict(data_norm_poly[folds==1]))
```

What is the $R^2$ performance on the train and test set of `data_norm_poly` for $d=2$?

```python
d=2

polynomial_features = PolynomialFeatures(degree=d)
data_norm_poly=pd.DataFrame(polynomial_features.fit_transform(data_norm))

model.fit(data_norm_poly[folds==0],target[folds==0])

print "R-squared train = %f" % metrics.r2_score(target[folds==0],model.predict(data_norm_poly[folds==0]))
print "R-squared test = %f" % metrics.r2_score(target[folds==1],model.predict(data_norm_poly[folds==1]))
```

What is the $R^2$ performance on the train and test set of `data_norm_log2_poly` for $d=2$?

```python
data_norm_log2_poly=pd.DataFrame(polynomial_features.fit_transform(data_norm_log2))

model.fit(data_norm_log2_poly[folds==0],target[folds==0])

print "R-squared train = %f" % metrics.r2_score(target[folds==0],model.predict(data_norm_log2_poly[folds==0]))
print "R-squared test = %f" % metrics.r2_score(target[folds==1],model.predict(data_norm_log2_poly[folds==1]))
```
